```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set Dimensions and delta
lenX = 18
lenY = 18
lenZ = 18
delta = 1

# Boundary condition
Ttop = 30
Tbottom = 30
Tleft = 30
Tright = 30
Tfront = 30
Tback = 30

T = np.empty((lenX, lenY, lenZ))
T.fill(30)
t = np.empty((lenX, lenY))

# Set Boundary condition
T[ :, :, 0] = Ttop
T[ :, :,-18] = Tbottom        
T[ :, 0, :] = Tright     
T[ :,-18, :] = Tleft 
T[ 0, :, :] = Tfront  
#T[-18, :, :] = Tback       

# Set Air Conditioner Boundary 
AIRCON_Lstart = 1 * lenX // 6
AIRCON_Lend = 3 * lenX // 6
AIRCON_Wstart =4 * lenX // 6
AIRCON_Wend = 5 * lenX // 6
T[-18, AIRCON_Lstart:AIRCON_Lend, AIRCON_Wstart:AIRCON_Wend] = 16

# Function to plot temperature distribution
def plot_temperature_distribution(T, iteration, colorinterpolation=50, colourMap=plt.cm.jet):
    fig = plt.figure(dpi = 80)
    ax = fig.add_subplot(111, projection='3d')
    X, Y, Z = np.meshgrid(np.arange(0, lenX), np.arange(0, lenY), np.arange(0, lenZ))
    scat = ax.scatter(X, Y, Z, c=T.flatten(), cmap=colourMap)
    plt.title(f"Temperature distribution at iteration {iteration}")
    fig.colorbar(scat, ax=ax, shrink=1.5, aspect=5)
    plt.show()

# Iteration 
iterations_to_plot = [1 ,2 ,5 ,10 ,20 ,50 ,100 ,200 ,500 ,1000]
maxIter = 1000
for iteration in range(0, maxIter+1):
    for i in range(1, lenX-1, delta):
        for j in range(1, lenY-1, delta):
            for k in range(1, lenZ-1, delta):
                T[i, j, k] = 1/6 * (T[i+1, j, k] + T[i-1, j, k] + T[i, j+1, k] + T[i, j-1, k] + T[i, j, k+1] + T[i, j, k-1])   
    if iteration in iterations_to_plot:
        plot_temperature_distribution(T, iteration)
    if iteration == 1000 :
        for i in range(18):
            for j in range(18):
                t[i, j] = T[i, j, 12]
# X, Y plane at z=12        
X, Y = np.meshgrid(np.arange(0, lenX), np.arange(0, lenY)) 
fig = plt.figure(dpi = 80)
ax = fig.add_subplot(111)
plt.title("Contour of Temperature")
plt.contourf(X, Y, t, cmap=plt.cm.jet)
plt.colorbar()
plt.show()  

print("åœ¨åº§æ¨™(4, 6, 12)çš„æº«åº¦:" ,T[4, 6, 12])
```
